```python
reg = linear_model.LinearRegression()
```

```python
df = pd.read_csv('hiring.csv')
df.head()
```

```python
df.experience = df.experience.fillna('zero')
df.head()
```

```python
df.isnull()
```

```python
x1 = np.mean(df[['test_score(out of 10)']])
df[['test_score(out of 10)']] = df[['test_score(out of 10)']].fillna(x1)
df[['test_score(out of 10)']] = df[['test_score(out of 10)']].astype(int)
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
```

```python
df
```

```python
x = df.drop('salary($)', axis = 1)
y = df['salary($)']
```

```python
reg.fit(x, y)
```

```python
query = [[2, 9, 6], [12, 10, 10]]
reg.predict(query)
```
